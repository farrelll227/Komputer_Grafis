```python
"""
DATAPLOT4
"""

import matplotlib.pyplot as plt
import numpy as np

t = [0, 20, 40, 60, 80, 100, 120]     # Time
T = [28, 30, 35, 43, 55, 70, 85]        # Temperature
p = [1.8, 2.3, 3, 4, 5.4, 7.3, 9.6]  # Pressure

fig, ax1 = plt.subplots()      # ---- set up a plot ax1 with subplots

ax1.grid(True)                 # ---- draw grid

ax1.set_xlabel('Time (hrs)')  # ---- label x axis of ax1

line1 = ax1.plot(t, T, 'r', label='Temperature')  # ---- plot temperature in red as curve 1
ax1.set_ylim([0, 100])        # ---- set y axis limits of ax1
ax1.set_ylabel('Temperature (°K)', color='r')     # ---- label y axis of ax1

ax2 = ax1.twinx()             # ---- set up ax2 as twin of ax1

line2 = ax2.plot(t, p, 'b', label='Pressure')     # ---- plot pressure in blue as curve 2
ax2.set_ylim([0, 10])        # ---- set y axis limits of ax2
ax2.set_ylabel('Pressure (psi)', color='b')       # ---- label y axis of ax2

lines = line1 + line2
labels = [l.get_label() for l in lines]
ax1.legend(lines, labels, loc='upper left')       # ---- plot legend

plt.title('Test Data')

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Konfigurasi plot
plt.figure(figsize=(10, 6))
plt.axis([0, 140, 0, 100])
plt.axis("on")
plt.grid(True)

t = [20, 40, 60, 80, 100, 120]
T = [30, 35, 43, 55, 70, 85]
p = [2, 3, 4, 5.3, 7.3, 9.6]
v = [.6, .58, .54, .46, .35, .2]

pp = []
for i in range(0, len(p), 1):
    pp.append(p[i] * 10)

vv = []
for i in range(0, len(v), 1):
    vv.append(v[i] * 100)

plt.plot(t, T, color='r', label='Temperature', marker='o')
plt.plot(t, pp, color='b', label='Pressure', marker='s')
plt.plot(t, vv, color='g', label='Volume', marker='d')
plt.legend(loc='upper left')

for y in np.arange(0, 100 + 1, 20):
    a = y / 10
    a = str(a)
    plt.text(142, y, a, color='b')

plt.xlabel('Time (hrs)')
plt.ylabel('Temperature °K', color='r')
plt.text(151, 65, 'Pressure (psi)', rotation=90, color='b')

for y in np.arange(100, -1, -20):
    a = y / 100
    a = str(a)
    plt.text(162, y, a, color='g')
    plt.text(159, y + 2, '-', color='g')

for y in np.arange(1, 99, 3):
    plt.text(157, y, '-', color='g')

plt.text(170, 65, r'Volume (cm³)', rotation=90, color='g')

plt.title('Compression Test Results')


n = len(v)

c1 = np.sum(t) / n
c2 = np.sum(v) / n

a = np.multiply(v, t)
c3 = np.sum(a)
b = np.multiply(t, t)
c4 = np.sum(b)

A = (c3 - n * c1 * c2) / (c4 - n * c1 * c1)
B = c2 - A * c1

for tp in np.arange(t[0], t[5], 2):
    vp = A * tp + B
    vp = vp * 100
    plt.scatter(tp, vp, color='g', s=1)

sumee = 0
for i in range(len(t)):
    e = (v[i] - (A * t[i] + B)) ** 2
    sumee += e

rms = np.sqrt(sumee / n)

plt.text(60, 28, 'v=At+B', color='g')
plt.arrow(78, 30, 6, 6, head_length=3, head_width=1.5, color='g', linewidth=0.5)

vp1 = A * t[0] + B
vp1 = '%7.4f' % (vp1)
plt.text(2, 64, vp1, color='g')

vp2 = A * t[5] + B
vp2 = '%7.4f' % (vp2)
plt.text(122, 25, vp2, color='g')

Ap = '%7.5f' % (A)
plt.text(65, 18, 'A=', color='g')
plt.text(72, 18, Ap, color='g')

Bp = '%7.5f' % (B)
plt.text(65, 12, 'B=', color='g')
plt.text(72, 12, Bp, color='g')

rms = '%7.3f' % (rms)
plt.text(95.3, 3, 'RMS error=', color='g')
plt.text(123, 3, rms, color='g')

plt.show()
```
